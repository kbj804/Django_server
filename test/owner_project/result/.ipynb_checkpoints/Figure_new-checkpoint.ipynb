```python
dfs = d_figure[['주소', '상태', '가입일자', '최종불입일자', '은행', '상품금액',
       '잔여금액', '연체횟수', '성별', '이름번호', '가입연차', '나이', '가입나이', '해약나이', '진행률']]
dfs.columns
```

```python
# 정규화 z-Score
def z_score_normalize(value, mean, std):
    normalized = []
    normalized_num = (value - mean) / std
    normalized.append(normalized_num)
    return normalized[0]
```

```python
z_score_list=['가입일자', '최종불입일자', '상품금액','잔여금액', '연체횟수', '가입연차', '나이', '가입나이', '해약나이']
def auto_zscore(z_list):
    for column in z_list:
        mean = np.mean(list(dfs[column]))
        std = np.mean(list(dfs[column]))
        dfs[column] = dfs[column].apply(lambda x : z_score_normalize(x,mean,std))
        
auto_zscore(z_score_list)
```
